```python

# EBOLA PAPER SENSIVITY
import numpy as np
import matplotlib.pyplot as plt

data = [+0.0003445187429, +0.06914093033, +0.09941632883, +0.04319876193, +0.0004547238101, 0.7877892552,
        -0.3608511922 , +0.005638575238,  +0.02103287375, -0.6233905101, -0.05143975652, -0.006093299053,
        +0.5031683582,  -0.7877892550,    +0.02128937712, +0.2707413272, +0.008240994585]

Labels = ["$\\delta$", "$\\beta$", "$\\beta_1$", "$\\beta_2$",  "$\\beta_3$",
          "$\\beta_p$", "$\\gamma_1$", "$b_1$", "$\\eta$", "$\\xi$", "$d$", "b", "$\\eta_1$",
          "$\\alpha$", "$\\gamma$", "$\\sigma$", "$d_1$"]

# Round data to 3 decimal places
data_rounded = [round(val, 4) for val in data]

# Generate different colors for each bar
colors = plt.cm.viridis(np.linspace(0, 1, len(data_rounded)))

plt.figure(figsize=(8, 5))
plt.bar(Labels, data_rounded, align='center', color=colors)
plt.xticks(rotation=45, ha='right')  # Rotate labels for better visibility
for i in range(len(Labels)):
    plt.text(i, data_rounded[i], data_rounded[i], ha='center', rotation=90)  # Rotate text vertically

plt.xlabel('Parameter')
plt.ylabel('Sensitivity index, $S_p^{R_0}$')
plt.tight_layout()

plt.grid(True)
plt.savefig("Sen.pdf")
plt.show()
```
